## Results for Notch-Delta-Jagged model

Equations for cell 1:

\begin{equation}
\frac{dN}{dt} = N_0 H^{S+}(S) - k_c N (D+J)  - k_t N (D_{ext} + J_{ext}) - \gamma N
\end{equation}

\begin{equation}
\frac{dD}{dt} = D_0 H^{S-}(I) - k_c D N - k_t D N_{ext} - \gamma D
\end{equation}

\begin{equation}
\frac{dJ}{dt} = J_0 H^{S+}(I) - k_c J N - k_t J N_{ext} - \gamma J
\end{equation}

\begin{equation}
\frac{dI}{dt} = k_t N (D_{ext} + J_{ext}) - \gamma_I I 
\end{equation}



```python
import numpy as np
import random as rand
class NullDevice():
    def write(self, s):
        pass
original_stdout = sys.stdout

from PyDSTool import *
from PyDSTool.Toolbox import phaseplane as pp
import PyDSTool_aux_functions as aux
%load_ext autoreload
%autoreload 2
```

```python
nmodel = 'modelNDJ'
names       = {'N': 'Notch'
              ,'D': 'Dll'
              ,'J': 'Jag'
              ,'I': 'NICD'
              }
plt.rcParams.update({'font.size': 16}) 
DSargs      = args(name=nmodel, checklevel=2)
DSargs.pars = {'g'  : 1.0e-1                                   # degradation of a protein 
              ,'gI' : 5.0e-1                                   # degradation of NICD 
              ,'kc' : 5.0e-4                                   # rate of cis interaction
              ,'kt' : 5.0e-5                                   # rate of trans interaction
              ,'p'  : 2.0e+0, 'pj' : 5.0e+0                    # hill coeficient
              ,'s0' : 2.0e+2                                   # hill factor
              ,'Nt' : 0.0e+3, 'Lt' : 0.0e+3                    # number of external proteins
              ,'N0' : 1.6e+3, 'D0' : 1.8e+3, 'J0' : 1.2e+3     # production rate
              ,'ln' : 2.0e+0, 'ld' : 0.0e+0, 'lj' : 2.0e+0     # lambda of shifted hill function
              }
DSargs.varspecs = {'N': 'N0*HS(I,s0,p ,ln) - N*( kc*(D+J) + kt*Lt ) - g*N'
                  ,'D': 'D0*HS(I,s0,p ,ld) - D*( kc*N     + kt*Nt ) - g*D'
                  ,'J': 'J0*HS(I,s0,pj,lj) - J*( kc*N     + kt*Nt ) - g*J'
                  ,'I': 'kt*N*Lt - gI*I'
                  }
DSargs.fnspecs  = [aux.functions()]
DSargs.ics      = {'N': 0, 'D': 0, 'J': 0, 'I': 0
                  }
DSargs.xdomain  = {'N': [0, 2.0e+4], 'D': [0, 2.0e+4], 'J': [0, 2.0e+4], 'I': [0, 2.0e+4]
                  }
DSargs.pdomain  = {'Lt': [0, 2500]
                  }
DSargs.tdomain  = [0., 100.]
DSargs.algparams= {'init_step':0.1}
ODE = Vode_ODEsystem(DSargs)
```

```python
#################################################################################################
# ----------------------------------------- Figure 4D ------------------------------------------
#################################################################################################

freepar = 'Lt'
ODE.set(pars = {freepar: 1e-2, 'Nt': 3000} ) 
fp_coord = aux.eliminate_redundants(pp.find_fixedpoints(ODE, n=4, maxsearch=1e+4, eps=1e-10),6)
ODE.set(ics  = fp_coord[0])  
PCargs = aux.PyCont_args(nmodel, freepar, 150, saveeigen=True, maxstep=1e+2, minstep=1e-2, step=5e+1)              
lp = aux.plot_continuation(ODE, PCargs, nmodel,['N'], freepar, names, LimitPoints='LP', off_points=True, 
                           fontsize=20, xlim=[800,1400])
                        #   save_fig='./figures/BifurcationNDJ_N300.pdf', fontsize=20, xlim=[800,1400])
for i in range(4):
    plt.axvline(x=lp[i],color='k')
```
