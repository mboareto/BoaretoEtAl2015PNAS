## Results for Notch-Delta model

Equations:

\begin{equation}
\frac{dN}{dt} = N_0 H^{S+}(I) - k_c N D  - k_t N D_{ext} - \gamma N
\end{equation}

\begin{equation}
\frac{dD}{dt} = D_0 H^{S-}(I) - k_c D N - k_t D N_{ext} - \gamma D 
\end{equation}

\begin{equation}
\frac{dI}{dt} = k_t N D_{ext} - \gamma_I I 
\end{equation}

```python
import numpy as np
import matplotlib.pyplot as plt
import random as rand
class NullDevice():
    def write(self, s):
        pass
original_stdout = sys.stdout

from PyDSTool import *
from PyDSTool.Toolbox import phaseplane as pp
import PyDSTool_aux_functions as aux
%load_ext autoreload
%autoreload 2
```

```python
nmodel = 'modelND'
names  = {'N': 'Notch'
         ,'D': 'Delta'
         ,'I': 'NICD'
         }
plt.rcParams.update({'font.size': 16}) 
DSargs      = args(name=nmodel, checklevel=2)
DSargs.pars = {'g'  : 1.0e-1                                  # degradation of a protein 
              ,'gI' : 5.0e-1                                  # degradation of NICD 
              ,'kc' : 5.0e-4                                  # rate of cis interaction
              ,'kt' : 5.0e-5                                  # rate of trans interaction
              ,'p'  : 2.0e+0                                  # hill coeficient
              ,'s0' : 2.0e+2                                  # hill factor
              ,'Nt' : 1.0e+0, 'Dt' : 1.0e+0                   # number of external proteins
              ,'N0' : 5.0e+2, 'D0' : 1.0e+3                   # production rate of the proteins
              ,'ln' : 2.0e+0, 'ld' : 0.0e+0                   # lambda of shifted hill functions
              }
DSargs.varspecs = {'N': 'N0*HS(I,s0,p,ln) - N*( kc*D + kt*Dt ) - g*N'
                  ,'D': 'D0*HS(I,s0,p,ld) - D*( kc*N + kt*Nt ) - g*D'
                  ,'I': 'kt*N*Dt - gI*I'
                  }
DSargs.fnspecs  = [aux.functions()]
DSargs.ics      = {'N': 0, 'D': 0, 'I': 0
                  }
DSargs.xdomain  = {'N': [0, 2.0e+4], 'D': [0, 2.0e+4], 'I': [0, 5.0e+3]
                  }
DSargs.pdomain  = {'Dt': [0, 4000]
                  }
DSargs.tdomain  = [0., 500.]
DSargs.algparams= {'init_step':0.1}
ODE = Vode_ODEsystem(DSargs)
```

```python
#################################################################################################
# ----------------------------------------- Figure 3C ------------------------------------------
#################################################################################################

freepar='Dt'
DSargs.pdomain['Dt'] = [0,3500]
ODE = Vode_ODEsystem(DSargs)
ODE.set(pars = {freepar: 1e-0, 'Nt': 500} ) 
fp = aux.eliminate_redundants(pp.find_fixedpoints(ODE, n=4, maxsearch=1e+4, eps=1e-10),6)
ODE.set(ics  = fp[0])  
PCargs = aux.PyCont_args(nmodel, freepar, 200, saveeigen=True, maxstep=1e+2, minstep=1e-2, step=5e+1)              
lp = aux.plot_continuation(ODE, PCargs, nmodel,['N'], freepar, names, xmin=-1, LimitPoints='LP',
                           off_points=True, save_fig='./figures/BifurcationND_N.pdf', fontsize=20)
for i in range(2):
    plt.axvline(x=lp[i],color='k')       
#plt.savefig('./figures/BifurcationND_N.pdf', format='pdf', dpi=900)
```

```python
#################################################################################################
# ---------------------------------------- Figure S5A -------------------------------------------
#################################################################################################

freepar = 'Dt'
ODE.set(pars = {freepar: 1e-0, 'Nt': 500} ) 
fp = aux.eliminate_redundants(pp.find_fixedpoints(ODE, n=4, maxsearch=1e+4, eps=1e-10),6)
ODE.set(ics  = fp[0])  
PCargs = aux.PyCont_args(nmodel, freepar, 150, saveeigen=True, maxstep=1e+2, minstep=1e-2, step=5e+1)
aux.plot_continuation(ODE, PCargs, nmodel,['N','D','I'], freepar, names, xmin=-1, off_points=False, 
                      fontsize=20, normal_form_coef=True)   
                      #save_fig='./figures/BifurcationND.pdf', fontsize=20, normal_form_coef=True)
```

```python
#################################################################################################
# ----------------------------------------- Figure 3D ------------------------------------------
#################################################################################################

sys.stdout = NullDevice()
freepar='Dt'
r = range(0,6001,250)
x  = []

for Nt in r:
    ODE.set(pars = {'Nt': Nt, 'Dt': 1e-0} )
    fp = aux.eliminate_redundants(pp.find_fixedpoints(ODE, n=2, maxsearch=1e+4, eps=1e-12),4)
    ODE.set(ics  = fp[0])    
    PCargs = aux.PyCont_args(nmodel, freepar, 150, saveeigen=True, maxstep=1e+2, minstep=1e-2, step=5e+1) 
    lp = aux.plot_continuation(ODE, PCargs, nmodel,['N'], freepar, names, fs=[4,3],
                              off_points=True, LimitPoints='LP', fontsize=12, showcurve=False)
    for i in range(len(lp)):
        x += [[i,lp[i]]]
sys.stdout = original_stdout

figure(figsize=(6,5), dpi=500)
x = asarray(x)
for i in range(0,2):
     plt.plot(x[x[:,0]==i][:,1], r, color='k', linewidth=1.5)        
plt.xlabel('Dt', fontsize= 18)
plt.ylabel('Nt', fontsize= 18)
plt.ylim(0,6000)
plt.xlim(0,3500)
#plt.savefig('./figures/PhaseDiagramND.pdf', format='pdf', dpi=900)
```
