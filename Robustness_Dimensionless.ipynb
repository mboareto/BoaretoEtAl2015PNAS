## Results for Notch-Delta-Jagged model

Equations for cell 1:

\begin{equation}
\frac{dN}{dt} = N_0 H^{S+}(S) - k_c N (D+J)  - k_t N (D_{ext} + J_{ext}) - \gamma N
\end{equation}

\begin{equation}
\frac{dD}{dt} = D_0 H^{S-}(I) - k_c D N - k_t D N_{ext} - \gamma D
\end{equation}

\begin{equation}
\frac{dJ}{dt} = J_0 H^{S+}(I) - k_c J N - k_t J N_{ext} - \gamma J
\end{equation}

\begin{equation}
\frac{dI}{dt} = k_t N (D_{ext} + J_{ext}) - \gamma_I I 
\end{equation}

```python
nmodel = 'modelND'
names  = {'N': 'Notch'
         ,'D': 'Delta'
         ,'I': 'NICD'
         }
plt.rcParams.update({'font.size': 16}) 
DSargs      = args(name=nmodel, checklevel=2)
DSargs.pars = {'g'  : 1.0e-1                                  # degradation of a protein 
              ,'gI' : 5.0e-1                                  # degradation of NICD 
              ,'kc' : 5.0e-4                                  # rate of cis interaction
              ,'kt' : 5.0e-5                                  # rate of trans interaction
              ,'p'  : 2.0e+0                                  # hill coeficient
              ,'s0' : 2.0e+2                                  # hill factor
              ,'Nt' : 1.0e+0, 'Dt' : 1.0e+0                   # number of external proteins
              ,'N0' : 5.0e+2, 'D0' : 1.0e+3                   # production rate of the proteins
              ,'ln' : 2.0e+0, 'ld' : 0.0e+0                   # lambda of shifted hill functions
              ,'tau': 0.0
              ,'aD' : 0.0
              ,'bD' : 0.0
              ,'aN' : 0.0
              ,'bN' : 0.0
              ,'bI': 0.0
              }
DSargs.varspecs = {'N': '( (1 + Hp(I,1,p)) - N*( aD*D + bD ) - N)/tau'
                  ,'D': '(  Hn(I,1,p)      - D*( aN*N + bN ) - D)/tau'
                  ,'I': 'bI*bD*N - I'
                  }
DSargs.fnspecs  = [aux.functions()]
DSargs.ics      = {'N': 0, 'D': 0, 'I': 0
                  }
DSargs.xdomain  = {'N': [0, 2.0e+4], 'D': [0, 2.0e+4], 'I': [0, 5.0e+3]
                  }
DSargs.pdomain  = {'bD': [0, 3]
                  }
DSargs.tdomain  = [0., 500.]
DSargs.algparams= {'init_step':0.1}
ODE = Vode_ODEsystem(DSargs)
```

```python
p = DSargs.pars
p['tau'] = p['gI']/p['g']
p['aD']  = p['kc']*p['D0']/(p['g']*p['g'])
p['aN']  = p['kc']*p['N0']/(p['g']*p['g'])
p['bD']  = p['kt']*p['Dt']/p['g']
p['bN']  = p['kt']*p['Nt']/p['g']
p['bI']  = p['N0']/(p['s0']*p['gI'])

ODE = Vode_ODEsystem(DSargs)

freepar = 'bD'
ODE.set(pars = {freepar: 0.1, 'Nt': 500} ) 
fp = aux.eliminate_redundants(pp.find_fixedpoints(ODE, n=4, maxsearch=1e+4, eps=1e-10),6)
fp
```

```python
ODE.set(ics  = fp[0])  
PCargs = aux.PyCont_args(nmodel, freepar, 250, saveeigen=True, maxstep=5e-2, minstep=1e-3, step=1e-2)
plt.rcParams.update({'font.size': 16}) 
aux.plot_continuation(ODE, PCargs, nmodel,['N','D','I'], freepar, names, xmin=-1, off_points=False, 
                      fontsize=20,  xlim=[0,3], normal_form_coef=False)
```

```python
DSargs.pars['Dt'] = 200
DSargs.pars['Nt'] = 200
p = DSargs.pars
p['tau'] = p['gI']/p['g']
p['aD']  = p['kc']*p['D0']/(p['g']*p['g'])
p['aN']  = p['kc']*p['N0']/(p['g']*p['g'])
p['bD']  = p['kt']*p['Dt']/p['g']
p['bN']  = p['kt']*p['Nt']/p['g']
p['bI']  = p['N0']/(p['s0']*p['gI'])
ODE = Vode_ODEsystem(DSargs)

aux.param_sensitivity_bars(['tau','bI','aD','aN','bD','bN','p'], ODE, DSargs, 'I',
                           save_fig='./figures/ParamSensitivity_DimlessND.pdf')
```
