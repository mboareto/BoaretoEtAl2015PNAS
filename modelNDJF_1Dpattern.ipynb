## Notch-Delta

Equations for cell 1:

\begin{equation}
\frac{dN}{dt} = \gamma_N(N_0 H^+(S) - N) - k_c N D  - k_t N D_t
\end{equation}

\begin{equation}
\frac{dD}{dt} = \gamma_D (D_0 H^-(S) - D) - k_c D N - k_t D N_t
\end{equation}

\begin{equation}
\frac{dS}{dt} = k_t N D_t - \gamma_S S 
\end{equation}

```python
import numpy as np
import timeit
from itertools import combinations
from random import uniform
from time import clock
plt.rcParams.update({'font.size': 22}) 

# ------------------------------------hill functions------------------------------------#
def HS(X,X0,nX,l):
    return l + (1.0-l)*Hn(X,X0,nX)
def Hn(X,X0,nX):
    return 1.0/(1.0 + (X/X0)**nX)
def Hp(X,X0,nX):
    return 1.0 - Hn(X,X0,nX)
#----------------------------------------------------------------------------------------#
#-----------------------sum proteins of the neighbors------------------------------------#
def sum_nsignal(p,pts,i,ncells,key):
    X  = np.zeros(ncells+2)
    X[ncells+1] = pts[i][0]
    X[0] = pts[i][ncells-1]
    X[1:(ncells+1)] = pts[i]
        
    if i=='N':
        I = np.zeros(ncells+2)
        I[ncells+1] = pts['I'][0]
        I[0] = pts['I'][ncells-1]
        I[1:(ncells+1)] = pts['I']
        if key=='D':
            return 0.5*(X[0:ncells]*eval('HS(I,s0,pf,ldf)',p,{'I': I[0:ncells]}) 
                        + X[2:ncells+2]*eval('HS(I,s0,pf,ldf)',p,{'I': I[2:ncells+2]}) )
        if key=='J':
            return 0.5*(X[0:ncells]*eval('HS(I,s0,pf,ljf)',p,{'I': I[0:ncells]}) 
                        + X[2:ncells+2]*eval('HS(I,s0,pf,ljf)',p,{'I': I[2:ncells+2]}) )
    else:
        return 0.5*(X[0:ncells] + X[2:ncells+2])
#----------------------------------------------------------------------------------------#
#-----------------------euler method for calculating trajectory--------------------------#    
def euler_traj(p,t,dt,ncells,v_lim):
    pts = {}   
    for j in vlim.keys():
        pts[j] = np.random.uniform(vlim[j][0],vlim[j][1],ncells)  # random initial condictions

    for k in range(int(t/dt)):
        pts_new = {}
        for j in vlim.keys():
            pts_new[j] = pts[j] + dt*dfdt(p,pts,j,ncells)
        pts = pts_new
    return pts
#----------------------------------------------------------------------------------------#
#-----------------------sum external proteins of the neighbors---------------------------#
def dfdt(p,pts,key,ncells):
    for i in pts.keys():
        p[i+'t'] = sum_nsignal(p,pts,i,ncells,key) 
    return eval(eqs[key[0]], p, pts)
#----------------------------------------------------------------------------------------#
```

```python
p = {'g'  : 1.0e-1                                   # degradation of a protein 
    ,'gI' : 5.0e-1                                   # degradation of NICD 
    ,'kc' : 5.0e-4                                   # rate of cis interaction
    ,'kt' : 2.5e-5                                   # rate of trans interaction
    ,'p'  : 2.0e+0, 'pj' : 5.0e+0                    # hill coeficient
    ,'s0' : 2.0e+2                                   # hill factor
    ,'N0' : 1.4e+3, 'D0' : 1.6e+3, 'J0' : 1.2e+3     # production rate
    ,'ln' : 2.0e+0, 'ld' : 0.0e+0, 'lj' : 2.0e+0     # lambda of shifted hill function
    ,'pf' : 1.0
    ,'ldf': 3.0,    'ljf': 0.3
    ,'HS': HS, 'Hp': Hp, 'Hn': Hn
    }
vlim={'N': [0,6000], 'D': [0,6000], 'J': [0,6000], 'I': [0,600]}

eqs = {'N': 'N0*HS(I,s0,p ,ln) - N*( (kc*D + kt*Dt)*HS(I,s0,pf,ldf) + (kc*J + kt*Jt)*HS(I,s0,pf,ljf) ) - g*N'
      ,'D': 'D0*HS(I,s0,p ,ld) - D*(  kc*N*HS(I,s0,pf,ldf) + kt*Nt ) - g*D'
      ,'J': 'J0*HS(I,s0,pj,lj) - J*(  kc*N*HS(I,s0,pf,ljf) + kt*Nt ) - g*J'
      ,'I': 'kt*N*( Dt*HS(I,s0,pf,ldf) + Jt*HS(I,s0,pf,ljf) ) - gI*I'
      }
```

```python
start = timeit.default_timer()
ncells = 100
ncomb = 100
dt = 0.1
t  = 100
    
p['D0'] = 1.6e+3
r = np.arange(600, 1801, 100)
y = np.zeros((len(r),3))
v = np.zeros((len(r),3))
for k in range(len(r)):
    p['J0'] = r[k]
    z = np.zeros((3, ncomb))
    for i in range(ncomb):
        pts = euler_traj(p,t,dt,ncells,vlim)
        x = pts['I']
        z[0][i] = sum(x<100)
        z[1][i] = ncells - sum(x<100) - sum(x>300)
        z[2][i] = sum(x>300)
    for j in range(3):
        y[k][j] = np.mean(z[j][:])/float(ncells)

figure(figsize=(7,5))  
plot(r,y[:,0],'-o', color = [0.36862745,  0.66666667,  0.98039216], linewidth=2, markersize=20, label='(S)')
plot(r,y[:,1],'-^', color = [0.74117647,  0.0745098 ,  0.81176471], linewidth=2, markersize=20, label='(S/R)')
plot(r,y[:,2],'-s', color = [0.76470588,  0.10980392,  0.10980392], linewidth=2, markersize=20, label='(R)')
plt.ylim([-0.03, 1.03])
plt.xlabel('J0')
plt.ylabel('f')
#plt.savefig('./figures/1DpatterningNDJF_fractionStates_J0.pdf', format='pdf', dpi=900)
print (timeit.default_timer() - start)/60.0, 'minutes'                      # in a laptop i7
```
