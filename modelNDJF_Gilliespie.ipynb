## Results for Notch-Delta-Jagged including Fringe effect

Equations:

\begin{equation}
\frac{dN}{dt} = N_0 H^{S+}(I) -  N ( (k_c D + kt D_{ext}) H^{S+}(I)] + (k_c J + k_t J_{ext}) H^{S-}(I)) - \gamma N
\end{equation}

\begin{equation}
\frac{dD}{dt} = D_0 H^{S-}(I) - k_c D N H^{S+}(I)- k_t D N_{ext} - \gamma D
\end{equation}

\begin{equation}
\frac{dJ}{dt} = J_0 H^{S+}(I) - k_c J N H^{S-}(I) - k_t J N_{ext} - \gamma J 
\end{equation}

\begin{equation}
\frac{dI}{dt} = k_t N (D_{ext} H^{S+}(I) + J_{ext} H^{S-}(I)) - \gamma_I I 
\end{equation}

```python
def HS(X,X0,nX,l):
    return l + (1.0-l)*Hn(X,X0,nX)
def Hn(X,X0,nX):
    return 1.0/(1.0 + (X/X0)**nX)
def Hp(X,X0,nX):
    return 1.0 - Hn(X,X0,nX)
#--------------------------------------------------------------------------------------------------------------------#
#--------------------------------------------------------------------------------------------------------------------#
def gilliespie(N, D, J, I, t, p):
    a = [p['kt']*N*p['Dt']*HS(I,p['s0'],p['pf'],p['ldf']) + p['kt']*N*p['Jt']*HS(I,p['s0'],p['pf'],p['ljf'])   # a[0]
        ,p['kc']*N*D*HS(I,p['s0'],p['pf'],p['ldf'])                                                            # a[1]
        ,p['kc']*N*J*HS(I,p['s0'],p['pf'],p['ljf'])                                                            # a[2]
        ,p['N0']*HS(I,p['s0'],p['p'],p['ln'])                                                                  # a[3]
        ,p['D0']*HS(I,p['s0'],p['p'],p['ld'])                                                                  # a[4]
        ,p['J0']*HS(I,p['s0'],p['pj'],p['lj'])                                                                 # a[5]
        ,p['g']*N                                                                                              # a[6]
        ,p['kt']*D*p['Nt'] + p['g']*D                                                                          # a[7]
        ,p['kt']*J*p['Nt'] + p['g']*J                                                                          # a[8]
        ,p['gI']*I]                                                                                            # a[9]
    a = np.asarray(a)
    s = np.sum(a)

    r = rand.random()
    x = np.cumsum(a/s)
    if r < x[0]:
        N -= 1
        I += 1
    elif r < x[1]:
        N -= 1
        D -= 1
    elif r < x[2]:
        N -= 1
        J -= 1
    elif r < x[3]:
        N += 1
    elif r < x[4]:
        D += 1
    elif r < x[5]:
        J += 1
    elif r < x[6]:
        N -= 1
    elif r < x[7]:
        D -= 1
    elif r < x[8]:
        J -= 1
    elif r < x[9]:
        I -= 1
    #print x
    return N, D, J, I, t-log(rand.random())/s
#--------------------------------------------------------------------------------------------------------------------#
#--------------------------------------------------------------------------------------------------------------------#
def pts_gilliespie(N, D, J, I, t, p):
    pts = {}
    for i in ['N', 'D', 'J', 'I', 't']:
        pts[i] = []
    
    ts = 0
    while t < p['tdomain'][1]:
        [N,D,J,I,t] = gilliespie(N,D,J,I,t,p)
        if t > ts:
            pts['N'] += [N]
            pts['D'] += [D]
            pts['J'] += [J]
            pts['I'] += [I]
            pts['t'] += [t]
            ts += p['dt']
        
    return pts
#--------------------------------------------------------------------------------------------------------------------#
#--------------------------------------------------------------------------------------------------------------------#
p = {'g'  : 1.0e-1                                   # degradation of a protein 
    ,'gI' : 5.0e-1                                   # degradation of NICD 
    ,'kc' : 5.0e-4                                   # rate of cis interaction
    ,'kt' : 5.0e-5                                   # rate of trans interaction
    ,'p'  : 2.0e+0, 'pj' : 5.0e+0                    # hill coeficient
    ,'s0' : 2.0e+2                                   # hill factor
    ,'Nt' : 0.0e+3, 'Dt' : 0.0e+3, 'Jt' : 0.0e+3     # external proteins
    ,'N0' : 1.4e+3, 'D0' : 1.6e+3, 'J0' : 1.2e+3     # production rate
    ,'ln' : 2.0e+0, 'ld' : 0.0e+0, 'lj' : 2.0e+0     # lambda of shifted hill function
    ,'pf' : 1.0
    ,'ldf': 3.0,    'ljf': 0.3
    }

p['Dt'] = 600
p['Nt'] = 500
p['Jt'] = 2000
fp = [{'D': 3768.30217, 'I': 78.813,    'J': 4859.26181, 'N': 309.86466},   #  (S)
      {'D': 765.5319,   'I': 211.78611, 'J': 4756.19963, 'N': 848.11542},   # (S/R)
      {'D': 55.64883,   'I': 579.21728, 'J': 3461.26357, 'N': 2362.86601}]  #  (R)
```
